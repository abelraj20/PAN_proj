```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def func(x, t):
    dxdt = t - x * x
    return dxdt

h = 0.05
tmax = 9
t0 = 0
x0 = 1

xarr = []
tarr = []

xarr.append(x0)
tarr.append(t0)

while tarr[-1] < tmax:
    xnew = xarr[-1] + h * func(xarr[-1], tarr[-1])
    xarr.append(xnew)
    tarr.append(tarr[-1] + h)

print(xarr)
print(tarr)
```

```python
plt.plot(tarr, xarr)
plt.xlabel('t')
plt.ylabel('x(t)')
plt.show()
```
